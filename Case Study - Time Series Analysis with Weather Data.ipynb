# Data Cleaning

```python
rainfall_df = rainfall_df.drop(rainfall_df.iloc[:, 14:].columns, axis=1)
rainfall_df.head()
```

As before, it would be more appropriate, if we had the Date i.e., Year-Month as indices and the states as columns.

For this, we need to reshape the DataFrame as below.

```python
rainfall_df = rainfall_df.groupby(['YEAR', 'SUBDIVISION']).sum()
rainfall_df.head()
```

```python
rainfall_df = rainfall_df.stack()
rainfall_df.head()
```

```python
rainfall_df = rainfall_df.reset_index()
rainfall_df.rename(columns={'level_2': 'MONTH', 0 : 'RAINFALL in mm'}, inplace=True) # Rename the newly created columns.
rainfall_df.head()
```

```python
rainfall_df['MONTH'] = rainfall_df['MONTH'].str.capitalize() # Clean the data so that it can be converted to a DateTimeIndex
rainfall_df.head()
```

```python
from pandas.tseries.offsets import MonthEnd    # MonthEnd gives the last date of the given month

rainfall_df['Date'] = pd.to_datetime(rainfall_df['YEAR'].astype(str) + rainfall_df['MONTH'], format="%Y%b") + MonthEnd(1)
rainfall_df.head()
```

Since `YEAR` and `MONTH` are now redundant, let's drop them and set Date as the index as it is Unique.

```python
rainfall_df = rainfall_df.drop(['YEAR', 'MONTH'], axis=1)
rainfall_df.head()
```

Now, we have the Date as the Index, but we still don't have the states as columns. For that we will use Pivoting.
Pivoting is one of the most common operations and a must for a Data Analyst/Scientist know of.

```python
rainfall_df = rainfall_df.pivot(index='Date', columns='SUBDIVISION', values='RAINFALL in mm')
```

```python
moving_mean = rainfall_df.rolling(12).mean()    # Moving Averages are used regularly for Weather Plots and also Stock Market Analysis.
moving_mean.head()                              # Here, we have used a 12 month Moving Average. 
                                                # Hence the values will not be available for the first 12 months!
```

Now, it's time to visualize the plot.

Similar to Moving Averages, we'll use Moving Sum to find the years with highest rainfall.

```python
start_year = 1980
end_year = 2000
state = 'BIHAR'
plot_df = rainfall_df.loc[
                rainfall_df.index.year.isin(range(start_year, end_year)), 
                state]

plot_df.plot(linestyle='--',
                              legend=True, 
                              label='Monthly Rainfall')
plot_df.rolling(3).mean().plot(lw=3, 
                               legend=True, 
                               label='Moving Sum over 3 months of Monthly Rainfall'
                                                )

plot_df.rolling(6).mean().plot(
                            lw=3, 
                            legend=True, 
                            label='Moving Sum over 6 months of Monthly Rainfall',
                            grid=True,
                            )

_ = plt.title(
    'A Time Series Plot of the Monthly Rainfall from ' + str(start_year) + '-' + str(end_year) + ' in ' + state, 
    fontsize=30, 
    color='gold', 
    weight='bold', 
    style='italic',
            )
_ = plt.xlabel(
    'Year', 
    fontsize=25, 
    weight='bold', 
    color='darkgreen', 
    style='italic'
            )
_ = plt.ylabel(
    'Rainfall in mm', 
    fontsize=25, 
    weight='bold', 
    color='darkgreen', 
    style='italic'
            )
```
